```python
import cv2
import numpy as np
from ultralytics import YOLO

# --- C·∫§U H√åNH ƒê∆Ø·ªúNG D·∫™N ---
# H√£y thay ƒë·ªïi ƒë∆∞·ªùng d·∫´n n√†y cho ƒë√∫ng v·ªõi m√°y c·ªßa b·∫°n
model_path = r"C:\Users\tabao\OneDrive\Desktop\Vitex\test_keypoint\yolo26m-pose (1).pt"
image_path = r"C:\Users\tabao\OneDrive\Desktop\Vitex\test_keypoint\image.png"
# --------------------------

# 1. Load model
print("ƒêang load model...")
try:
    model = YOLO(model_path)
except Exception as e:
    print(f"L·ªói kh√¥ng t√¨m th·∫•y model t·∫°i: {model_path}")
    exit()

# 2. Run inference
print(f"ƒêang x·ª≠ l√Ω ·∫£nh: {image_path}...")
results = model(image_path)

# L·∫•y k·∫øt qu·∫£ c·ªßa ·∫£nh ƒë·∫ßu ti√™n
result = results[0]

# --- PH·∫¶N 1: IN TO·∫† ƒê·ªò RA M√ÄN H√åNH ---
print("\n--- K·∫æT QU·∫¢ T·ªåA ƒê·ªò KH·ªöP ---")
if result.keypoints is not None and len(result.keypoints) > 0:
    # L·∫•y tensor t·ªça ƒë·ªô x,y v√† chuy·ªÉn sang numpy (CPU)
    keypoints_numpy = result.keypoints.xy.cpu().numpy()
    
    for i, kps in enumerate(keypoints_numpy):
        print(f"\n>>> Ng∆∞·ªùi th·ª© {i+1}:")
        # kps l√† m·ªôt m·∫£ng (17, 2) ch·ª©a 17 kh·ªõp, m·ªói kh·ªõp c√≥ t·ªça ƒë·ªô [x, y]
        for j, (x, y) in enumerate(kps):
             # N·∫øu x, y ƒë·ªÅu b·∫±ng 0 c√≥ nghƒ©a l√† kh·ªõp ƒë√≥ kh√¥ng ƒë∆∞·ª£c ph√°t hi·ªán ho·∫∑c b·ªã che khu·∫•t
            status = "Ph√°t hi·ªán" if x > 0 and y > 0 else "Kh√¥ng r√µ/B·ªã che"
            print(f"  - Kh·ªõp {j:2d}: [{x:8.2f}, {y:8.2f}] ({status})")
else:
    print("Kh√¥ng t√¨m th·∫•y ng∆∞·ªùi/kh·ªõp n√†o trong ·∫£nh.")


# --- PH·∫¶N 2: HI·ªÇN TH·ªä ·∫¢NH K·∫æT QU·∫¢ ---
print("\n--- ƒêANG HI·ªÇN TH·ªä ·∫¢NH... ---")

# S·ª≠ d·ª•ng h√†m plot() t√≠ch h·ª£p s·∫µn c·ªßa YOLOv8 ƒë·ªÉ v·∫Ω k·∫øt qu·∫£
# B·∫°n c√≥ th·ªÉ th√™m tham s·ªë boxes=False v√†o trong plot() n·∫øu ch·ªâ mu·ªën hi·ªán x∆∞∆°ng m√† kh√¥ng hi·ªán h·ªôp bao quanh ng∆∞·ªùi
annotated_frame = result.plot()

# Hi·ªÉn th·ªã ·∫£nh d√πng OpenCV
cv2.imshow("YOLOv8 Pose Estimation Result", annotated_frame)

# (T√πy ch·ªçn) L∆∞u ·∫£nh k·∫øt qu·∫£ ra file
# output_path = image_path.replace(".png", "_result.png")
# cv2.imwrite(output_path, annotated_frame)
# print(f"ƒê√£ l∆∞u ·∫£nh k·∫øt qu·∫£ t·∫°i: {output_path}")

print("Nh·∫•n ph√≠m b·∫•t k·ª≥ tr√™n c·ª≠a s·ªï ·∫£nh ƒë·ªÉ tho√°t.")
# Gi·ªØ c·ª≠a s·ªï hi·ªÉn th·ªã cho ƒë·∫øn khi nh·∫•n ph√≠m b·∫•t k·ª≥
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# G√≥c khu·ª∑u tay (Elbow Angle): D√πng 3 ƒëi·ªÉm Vai - Khu·ª∑u - C·ªï tay.

1. Tay ph·∫£i: T√≠nh g√≥c gi·ªØa 3 ƒëi·ªÉm 6 - 8 - 10.

2. Tay tr√°i: T√≠nh g√≥c gi·ªØa 3 ƒëi·ªÉm 5 - 7 - 9.


# G√≥c n√°ch (Shoulder Angle): D√πng 3 ƒëi·ªÉm H√¥ng - Vai - Khu·ª∑u.

1. Tay ph·∫£i: T√≠nh g√≥c gi·ªØa 12 - 6 - 8.

2. Tay tr√°i: T√≠nh g√≥c gi·ªØa 11 - 5 - 7.

# ƒê·ªô khu·ªµu g·ªëi (Knee Flexion): D√πng 3 ƒëi·ªÉm H√¥ng - G·ªëi - C·ªï ch√¢n.

1. Ch√¢n ph·∫£i: T√≠nh g√≥c gi·ªØa 12 - 14 - 16.

2. Ch√¢n tr√°i: T√≠nh g√≥c gi·ªØa 11 - 13 - 15.

```python
import cv2
import numpy as np
from ultralytics import YOLO

def calculate_angle(a, b, c):
    """
    T√≠nh g√≥c gi·ªØa ba ƒëi·ªÉm a, b, c (v·ªõi b l√† ƒëi·ªÉm gi·ªØa).
    a, b, c l√† c√°c tuple ho·∫∑c list to·∫° ƒë·ªô [x, y].
    """
    a = np.array(a) # ƒêi·ªÉm ƒë·∫ßu (v√≠ d·ª•: Vai)
    b = np.array(b) # ƒêi·ªÉm gi·ªØa/Tr·ª•c (v√≠ d·ª•: Khu·ª∑u tay)
    c = np.array(c) # ƒêi·ªÉm cu·ªëi (v√≠ d·ª•: C·ªï tay)
    
    # T√≠nh g√≥c theo radian d√πng arctan2
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    
    # Chuy·ªÉn sang ƒë·ªô
    angle = np.abs(radians * 180.0 / np.pi)
    
    # Chu·∫©n h√≥a g√≥c v·ªÅ kho·∫£ng [0, 180] (v√¨ kh·ªõp ng∆∞·ªùi kh√¥ng b·∫ª ng∆∞·ª£c qu√° 180)
    if angle > 180.0:
        angle = 360 - angle
        
    return angle

# --- C·∫§U H√åNH ---
model_path = r"C:\Users\tabao\OneDrive\Desktop\Vitex\test_keypoint\yolo26m-pose (1).pt" 
image_path = r"C:\Users\tabao\OneDrive\Desktop\Vitex\test_keypoint\test2.png"

# Load model
model = YOLO(model_path)
results = model(image_path)
result = results[0] # L·∫•y ·∫£nh ƒë·∫ßu ti√™n

# Load ·∫£nh g·ªëc ƒë·ªÉ v·∫Ω th√¥ng s·ªë l√™n
image = cv2.imread(image_path)

if result.keypoints is not None and len(result.keypoints) > 0:
    # L·∫•y to·∫° ƒë·ªô c√°c kh·ªõp (chuy·ªÉn v·ªÅ CPU v√† numpy)
    # Shape: (17, 2)
    kps = result.keypoints.xy[0].cpu().numpy()

    # --- ƒê·ªäNH NGHƒ®A C√ÅC C·∫∂P 3 ƒêI·ªÇM (THEO Y√äU C·∫¶U C·ª¶A B·∫†N) ---
    # C·∫•u tr√∫c: "T√™n G√≥c": [ƒêi·ªÉm 1, ƒêi·ªÉm Gi·ªØa, ƒêi·ªÉm 3]
    # Index d·ª±a tr√™n b·∫£ng COCO (0-16)
    angles_config = {
        "Right Elbow": [6, 8, 10],   # Vai ph·∫£i - Khu·ª∑u ph·∫£i - C·ªï tay ph·∫£i
        "Left Elbow":  [5, 7, 9],    # Vai tr√°i - Khu·ª∑u tr√°i - C·ªï tay tr√°i
        "Right Shoulder": [12, 6, 8],# H√¥ng ph·∫£i - Vai ph·∫£i - Khu·ª∑u ph·∫£i
        "Left Shoulder":  [11, 5, 7],# H√¥ng tr√°i - Vai tr√°i - Khu·ª∑u tr√°i
        "Right Knee": [12, 14, 16],  # H√¥ng ph·∫£i - G·ªëi ph·∫£i - C·ªï ch√¢n ph·∫£i
        "Left Knee":  [11, 13, 15]   # H√¥ng tr√°i - G·ªëi tr√°i - C·ªï ch√¢n tr√°i
    }

    print("-" * 30)
    print(f"{'G√ìC KH·ªöP':<20} | {'GI√Å TR·ªä (ƒê·ªò)':<10}")
    print("-" * 30)

    for name, points in angles_config.items():
        p1_idx, p2_idx, p3_idx = points
        
        # L·∫•y to·∫° ƒë·ªô x,y c·ªßa 3 ƒëi·ªÉm
        p1 = kps[p1_idx]
        p2 = kps[p2_idx]
        p3 = kps[p3_idx]

        # Ki·ªÉm tra n·∫øu to·∫° ƒë·ªô tin c·∫≠y (kh√°c 0,0)
        # (YOLO tr·∫£ v·ªÅ 0,0 n·∫øu kh√¥ng detect ƒë∆∞·ª£c kh·ªõp)
        if np.all(p1 > 0) and np.all(p2 > 0) and np.all(p3 > 0):
            # T√≠nh g√≥c
            angle = calculate_angle(p1, p2, p3)
            
            # 1. In ra m√†n h√¨nh console
            print(f"{name:<20} | {int(angle)}¬∞")

            # 2. V·∫Ω l√™n ·∫£nh (Visualize)
            # V·∫Ω s·ªë ƒëo g√≥c ngay t·∫°i v·ªã tr√≠ kh·ªõp gi·ªØa (p2)
            cv2.putText(image, str(int(angle)), 
                        (int(p2[0]), int(p2[1])), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 2, cv2.LINE_AA)
            
            # V·∫Ω c√°c ƒë∆∞·ªùng n·ªëi ƒë·ªÉ d·ªÖ nh√¨n (Optional)
            cv2.line(image, (int(p1[0]), int(p1[1])), (int(p2[0]), int(p2[1])), (0, 255, 0), 2)
            cv2.line(image, (int(p2[0]), int(p2[1])), (int(p3[0]), int(p3[1])), (0, 255, 0), 2)
            
        else:
            print(f"{name:<20} | Kh√¥ng ƒë·ªß ƒëi·ªÉm")

    # Hi·ªÉn th·ªã ·∫£nh
    cv2.imshow("Angle Analysis", image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

else:
    print("Kh√¥ng t√¨m th·∫•y keypoints n√†o.")
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
import json
import os

# --- H√ÄM T√çNH G√ìC ---
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    if angle > 180.0: angle = 360 - angle
    return angle

# --- C·∫§U H√åNH ---
input_video_path = r"C:\Users\tabao\OneDrive\Desktop\Vitex\test_keypoint\short_pickle.mp4"
output_video_path = r"C:\Users\tabao\OneDrive\Desktop\Vitex\test_keypoint\pickleball_analyzed.mp4"
output_json_path = r"C:\Users\tabao\OneDrive\Desktop\Vitex\test_keypoint\pickleball_data.json"
model_path = r"C:\Users\tabao\OneDrive\Desktop\Vitex\test_keypoint\yolo26m-pose (1).pt"
TARGET_FPS = 25  # Gi·∫£m xu·ªëng c√≤n 25fps ƒë·ªÉ nh·∫π m√°y

# --- KH·ªûI T·∫†O ---
print("ƒêang load model YOLO...")
model = YOLO(model_path)

cap = cv2.VideoCapture(input_video_path)
if not cap.isOpened():
    print(f"Kh√¥ng th·ªÉ m·ªü video: {input_video_path}")
    exit()

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
original_fps = cap.get(cv2.CAP_PROP_FPS) # FPS th·ª±c t·∫ø c·ªßa video g·ªëc
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Video ƒë·∫ßu ra set c·ª©ng 25 fps
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, TARGET_FPS, (width, height))

angles_config = {
    "Right Elbow": [6, 8, 10],   "Left Elbow": [5, 7, 9],
    "Right Shoulder": [12, 6, 8],"Left Shoulder": [11, 5, 7],
    "Right Knee": [12, 14, 16],  "Left Knee": [11, 13, 15]
}

dataset_records = []
print(f"B·∫Øt ƒë·∫ßu x·ª≠ l√Ω... FPS g·ªëc: {original_fps} -> Output: {TARGET_FPS}")

frame_count = 0      # ƒê·∫øm frame g·ªëc
processed_count = 0  # ƒê·∫øm frame ƒë√£ l∆∞u

while True:
    ret, frame = cap.read()
    if not ret: break 

    # Thu·∫≠t to√°n ch·ªçn l·ªçc frame ƒë·ªÉ gi·ªØ ƒë√∫ng t·ªëc ƒë·ªô 25fps
    expected_frame_index = processed_count * (original_fps / TARGET_FPS)
    
    if frame_count >= int(expected_frame_index):
        # ==> X·ª¨ L√ù FRAME N√ÄY
        
        # [QUAN TR·ªåNG] T√≠nh timestamp d·ª±a tr√™n frame g·ªëc ƒë·ªÉ ch√≠nh x√°c tuy·ªát ƒë·ªëi
        real_timestamp = frame_count / original_fps

        results = model.track(frame, persist=True, verbose=False)
        result = results[0]
        
        frame_data = {
            "frame": processed_count + 1,
            "timestamp": round(real_timestamp, 3), # <--- CH√çNH L√Ä C√ÅI B·∫†N C·∫¶N
            "angles": {},
            "keypoints": {}
        }

        if result.keypoints is not None and len(result.keypoints) > 0:
            kps = result.keypoints.xy[0].cpu().numpy()
            for name, points in angles_config.items():
                p1_idx, p2_idx, p3_idx = points
                p1, p2, p3 = kps[p1_idx], kps[p2_idx], kps[p3_idx]
                if np.all(p1 > 0) and np.all(p2 > 0) and np.all(p3 > 0):
                    angle = calculate_angle(p1, p2, p3)
                    frame_data["angles"][name] = float(round(angle, 2))
                    frame_data["keypoints"][name] = {"x": float(round(p2[0], 2)), "y": float(round(p2[1], 2))}
                    
                    # V·∫Ω l√™n h√¨nh
                    cv2.putText(frame, str(int(angle)), (int(p2[0]), int(p2[1]) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 255), 2, cv2.LINE_AA)
                    cv2.line(frame, (int(p1[0]), int(p1[1])), (int(p2[0]), int(p2[1])), (0, 255, 0), 2)
                    cv2.line(frame, (int(p2[0]), int(p2[1])), (int(p3[0]), int(p3[1])), (0, 255, 0), 2)

        dataset_records.append(frame_data)
        out.write(frame)
        cv2.imshow('Pickleball Analysis', frame)
        processed_count += 1
    
    frame_count += 1
    if cv2.waitKey(1) & 0xFF == ord('q'): break
    if frame_count % 50 == 0: print(f"Processing: {frame_count}/{total_frames}...")

cap.release()
out.release()
cv2.destroyAllWindows()

with open(output_json_path, 'w', encoding='utf-8') as f:
    json.dump(dataset_records, f, indent=4, ensure_ascii=False)

print(f"‚úÖ ƒê√£ xong! File JSON chu·∫©n th·ªùi gian t·∫°i: {output_json_path}")
```

N·∫øu bi·ªÉu ƒë·ªì c·ªßa b·∫°n th·∫•y g√≥c nh·∫£y m√∫a lung tung $\rightarrow$ Ng∆∞·ªùi ch∆°i ƒëang "v·∫©y" c·ªï tay ho·∫∑c gi·∫≠t c√πi ch·ªè (L·ªói k·ªπ thu·∫≠t).

# H√†m chia ƒë·ªông t√°c

```python
import cv2
import numpy as np
from ultralytics import YOLO
import json
import os

# ==========================================
# 1. C·∫§U H√åNH ƒê∆Ø·ªúNG D·∫™N & TH√îNG S·ªê
# ==========================================
input_video_path = r"C:\Users\tabao\OneDrive\Desktop\Vitex\test_keypoint\short_pickle.mp4"
output_video_path = r"C:\Users\tabao\OneDrive\Desktop\Vitex\test_keypoint\pickleball_analyzed.mp4"
# C√°c file json nh·ªè s·∫Ω ƒë∆∞·ª£c l∆∞u c√πng th∆∞ m·ª•c v·ªõi ƒë∆∞·ªùng d·∫´n n√†y:
base_output_path = r"C:\Users\tabao\OneDrive\Desktop\Vitex\test_keypoint" 

model_path = r"C:\Users\tabao\OneDrive\Desktop\Vitex\test_keypoint\yolo26m-pose (1).pt"

TARGET_FPS = 25          # Gi·ªØ video ƒë·∫ßu ra ·ªü m·ª©c 25fps
VELOCITY_THRESHOLD = 2.0 # (Pixel/frame) Ng∆∞·ª°ng t·ªëc ƒë·ªô ƒë·ªÉ b·∫Øt ƒë·∫ßu ghi h√¨nh c√∫ ƒë√°nh
COOLDOWN_FRAMES = 3     # S·ªë frame tƒ©nh ƒë·ªÉ x√°c nh·∫≠n k·∫øt th√∫c c√∫ ƒë√°nh

# ==========================================
# 2. CLASS X·ª¨ L√ù LOGIC T√ÅCH ƒê·ªòNG T√ÅC
# ==========================================
class ActionSplitter:
    def __init__(self, base_path):
        self.base_path = base_path
        self.prev_wrist_pos = None
        self.is_moving = False
        self.cooldown_counter = 0
        self.current_buffer = []
        self.move_count = 0
        
        # ƒê·ªãnh nghƒ©a ti√™u chu·∫©n ph√¢n lo·∫°i
        self.criteria = {
            "Serve": {"min_elbow": 150, "zone": "low"},
            "Dink": {"max_knee": 150, "max_vel": 10},
            "Backhand": {"pos": "cross"}
        }

    def classify(self, frames):
        """Ph√¢n lo·∫°i ƒë·ªông t√°c d·ª±a tr√™n d·ªØ li·ªáu g√≥c v√† v·ªã tr√≠"""
        if not frames: return "Unknown"
        
        # T√¨m frame c√≥ v·∫≠n t·ªëc l·ªõn nh·∫•t (Impact phase)
        sorted_frames = sorted(frames, key=lambda x: x['velocity'], reverse=True)
        impact = sorted_frames[0]
        
        kps = impact['keypoints']
        angles = impact['angles']
        
        # L·∫•y d·ªØ li·ªáu an to√†n
        def get_pt(name): return kps.get(name, {'x': 0, 'y': 0})
        
        wrist = get_pt('Right Wrist')
        shoulder = get_pt('Right Shoulder')
        hip = get_pt('Right Hip')
        
        elbow_ang = angles.get('Right Elbow', 0)
        knee_ang = angles.get('Right Knee', 180)
        
        avg_vel = sum(f['velocity'] for f in frames) / len(frames)

        # --- LOGIC PH√ÇN LO·∫†I ---
        # 1. Backhand: Tay ph·∫£i n·∫±m b√™n tr√°i vai ph·∫£i
        if wrist['x'] < shoulder['x']:
            return "Backhand"
        
        # 2. Serve: Tay th·∫≥ng + ƒê√°nh th·∫•p (Y l·ªõn) + M·∫°nh
        if wrist['y'] > hip['y'] and elbow_ang > self.criteria["Serve"]["min_elbow"] and avg_vel > 5.0:
            return "Serve"
            
        # 3. Dink: G·ªëi khu·ªµu + ƒê√°nh nh·∫π
        if knee_ang < self.criteria["Dink"]["max_knee"] and avg_vel < self.criteria["Dink"]["max_vel"]:
            return "Dink"
            
        # 4. Forehand (M·∫∑c ƒë·ªãnh)
        chest_y = (shoulder['y'] + hip['y']) / 2
        if wrist['y'] < chest_y:
            return "Forehand_Volley"
        
        return "Forehand_Drive"

    def save_file(self):
        """L∆∞u buffer hi·ªán t·∫°i ra file JSON ri√™ng bi·ªát"""
        if len(self.current_buffer) > 5: # Ch·ªâ l∆∞u n·∫øu c√∫ ƒë√°nh d√†i h∆°n 5 frame
            self.move_count += 1
            move_type = self.classify(self.current_buffer)
            
            start_t = self.current_buffer[0]['timestamp']
            end_t = self.current_buffer[-1]['timestamp']
            
            # T·∫°o t√™n file: action_1_Forehand.json
            filename = f"action_{self.move_count}_{move_type}.json"
            full_path = os.path.join(self.base_path, filename)
            
            data_to_save = {
                "id": self.move_count,
                "type": move_type,
                "duration": round(end_t - start_t, 2),
                "start_time": start_t,
                "end_time": end_t,
                "frames": self.current_buffer
            }
            
            with open(full_path, 'w', encoding='utf-8') as f:
                json.dump(data_to_save, f, indent=4)
                
            print(f"üíæ [ƒê√É L∆ØU] {filename} ({start_t}s -> {end_t}s)")
        
        # Reset sau khi l∆∞u
        self.current_buffer = []

    def process(self, frame_data):
        """M√°y tr·∫°ng th√°i: Nh·∫≠n di·ªán Start/Stop"""
        # 1. T√≠nh v·∫≠n t·ªëc c·ªï tay
        velocity = 0.0
        if 'Right Wrist' in frame_data['keypoints']:
            curr = np.array([frame_data['keypoints']['Right Wrist']['x'], frame_data['keypoints']['Right Wrist']['y']])
            if self.prev_wrist_pos is not None:
                velocity = np.linalg.norm(curr - self.prev_wrist_pos)
            self.prev_wrist_pos = curr
        
        frame_data['velocity'] = float(round(velocity, 2))

        # 2. Logic chuy·ªÉn ƒë·ªïi
        if velocity > VELOCITY_THRESHOLD:
            # ==> TR·∫†NG TH√ÅI: ƒêANG ƒê√ÅNH (ACTION)
            if not self.is_moving:
                self.is_moving = True
                print(f"üöÄ B·∫Øt ƒë·∫ßu ƒë·ªông t√°c m·ªõi t·∫°i {frame_data['timestamp']}s")
            
            self.current_buffer.append(frame_data)
            self.cooldown_counter = 0
            
        elif self.is_moving:
            # ==> TR·∫†NG TH√ÅI: CH·ªú (COOLDOWN)
            self.current_buffer.append(frame_data)
            self.cooldown_counter += 1
            
            # N·∫øu tay d·ª´ng qu√° l√¢u -> K·∫øt th√∫c
            if self.cooldown_counter > COOLDOWN_FRAMES:
                self.is_moving = False
                self.save_file()

# ==========================================
# 3. H√ÄM T√çNH G√ìC
# ==========================================
def calculate_angle(a, b, c):
    a, b, c = np.array(a), np.array(b), np.array(c)
    radians = np.arctan2(c[1] - b[1], c[0] - b[0]) - np.arctan2(a[1] - b[1], a[0] - b[0])
    angle = np.abs(radians * 180.0 / np.pi)
    if angle > 180.0: angle = 360 - angle
    return angle

# ==========================================
# 4. CH∆Ø∆†NG TR√åNH CH√çNH
# ==========================================
print("ƒêang load model YOLO...")
model = YOLO(model_path)

# Kh·ªüi t·∫°o b·ªô t√°ch ƒë·ªông t√°c
splitter = ActionSplitter(base_output_path)

cap = cv2.VideoCapture(input_video_path)
if not cap.isOpened():
    print(f"Kh√¥ng th·ªÉ m·ªü video: {input_video_path}")
    exit()

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
original_fps = cap.get(cv2.CAP_PROP_FPS)
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, TARGET_FPS, (width, height))

angles_config = {
    "Right Elbow": [6, 8, 10],   "Left Elbow": [5, 7, 9],
    "Right Shoulder": [12, 6, 8],"Left Shoulder": [11, 5, 7],
    "Right Knee": [12, 14, 16],  "Left Knee": [11, 13, 15]
}
kp_map = {0: "Nose", 6: "Right Shoulder", 8: "Right Elbow", 10: "Right Wrist", 12: "Right Hip", 14: "Right Knee"}

print(f"B·∫Øt ƒë·∫ßu x·ª≠ l√Ω... FPS g·ªëc: {original_fps} -> Output: {TARGET_FPS}")

frame_count = 0      
processed_count = 0  

while True:
    ret, frame = cap.read()
    if not ret: break 

    # Logic gi·∫£m FPS
    expected_frame_index = processed_count * (original_fps / TARGET_FPS)
    
    if frame_count >= int(expected_frame_index):
        # [QUAN TR·ªåNG] T√≠nh timestamp ch√≠nh x√°c
        real_timestamp = frame_count / original_fps

        results = model.track(frame, persist=True, verbose=False)
        
        if results[0].keypoints is not None and len(results[0].keypoints) > 0:
            kps = results[0].keypoints.xy[0].cpu().numpy()
            
            frame_data = {
                "frame": processed_count + 1,
                "timestamp": round(real_timestamp, 3),
                "angles": {},
                "keypoints": {}
            }

            # 1. Tr√≠ch xu·∫•t to·∫° ƒë·ªô (D√πng cho logic ph√¢n lo·∫°i)
            for idx, name in kp_map.items():
                if kps[idx][0] > 0:
                    frame_data['keypoints'][name] = {"x": float(kps[idx][0]), "y": float(kps[idx][1])}

            # 2. T√≠nh g√≥c v√† V·∫Ω
            for name, points in angles_config.items():
                p1, p2, p3 = kps[points[0]], kps[points[1]], kps[points[2]]
                if np.all(p1 > 0) and np.all(p2 > 0) and np.all(p3 > 0):
                    angle = calculate_angle(p1, p2, p3)
                    frame_data["angles"][name] = float(round(angle, 2))
                    
                    # V·∫Ω
                    cv2.putText(frame, str(int(angle)), (int(p2[0]), int(p2[1]) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 255), 2, cv2.LINE_AA)
                    cv2.line(frame, (int(p1[0]), int(p1[1])), (int(p2[0]), int(p2[1])), (0, 255, 0), 2)
                    cv2.line(frame, (int(p2[0]), int(p2[1])), (int(p3[0]), int(p3[1])), (0, 255, 0), 2)

            # 3. G·ª≠i d·ªØ li·ªáu v√†o b·ªô t√°ch file (QUAN TR·ªåNG)
            splitter.process(frame_data)

        out.write(frame)
        cv2.imshow('Pickleball Analysis', frame)
        processed_count += 1
    
    frame_count += 1
    if cv2.waitKey(1) & 0xFF == ord('q'): break
    if frame_count % 50 == 0: print(f"Processing: {frame_count}/{total_frames}...")

cap.release()
out.release()
cv2.destroyAllWindows()

print(f"‚úÖ Ho√†n t·∫•t! C√°c file JSON ƒë√£ ƒë∆∞·ª£c l∆∞u t·∫°i th∆∞ m·ª•c: {base_output_path}")
```
